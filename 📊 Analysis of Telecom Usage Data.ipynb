# Age Distribution of Users

```python
# Histogram of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, color='skyblue')
plt.title('Age Distribution of Users')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Gender Distribution 
# Countplot of Gender
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=df, palette='Set2')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
 # Primary Use of Mobile Phones
plt.figure(figsize=(10, 7))
df['Primary Use'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('coolwarm'))
plt.title('Primary Use of Mobile Phones')
plt.ylabel('')
plt.show()
```

# Usage Pattern Analysis

```python
avg_screen_time = df.groupby('Primary Use')['Screen Time (hrs/day)'].mean()
print(avg_screen_time)

# Bar plot
colors = ['orange', 'blue', 'green', 'red']

plt.figure(figsize=(10, 6))
avg_screen_time.plot(kind='bar', color=colors)
plt.title('Average Screen Time by Primary Use')
plt.xlabel('Primary Use')
plt.ylabel('Screen Time (hrs/day)')
plt.show()
```

#ðŸ”¹ Monthly Recharge Cost by Gender

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Monthly Recharge Cost (INR)', data=df, palette='Set3')
plt.title('Monthly Recharge Cost by Gender')
plt.xlabel('Gender')
plt.ylabel('Monthly Recharge Cost (INR)')
plt.show()
```

# Data Usage Across Different Locations

```python
avg_data_usage = df.groupby('Location')['Data Usage (GB/month)'].mean()
print(avg_data_usage)

# Bar plot
colors = ['red', 'blue', 'green', 'orange']

plt.figure(figsize=(12, 6))
avg_data_usage.sort_values(ascending=False).plot(kind='bar', color=colors)
plt.title('Average Data Usage by Location')
plt.xlabel('Location')
plt.ylabel('Data Usage (GB/month)')
plt.show()
```

# Correlation Analysis (Numerical Variables)

```python
# Select only numerical columns
numerical_df = df.select_dtypes(include=['number'])

# Plot the correlation heatmap for numerical columns only
plt.figure(figsize=(12, 8))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap for Numerical Columns')
plt.show()
```
